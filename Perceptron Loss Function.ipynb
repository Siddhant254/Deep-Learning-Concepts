```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# GENERATING CLASSIFICATION DATASET
from sklearn.datasets import make_classification
X,y = make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,
                          n_classes=2,n_clusters_per_class=1,random_state=41,hypercube=False,
                          class_sep=15)

plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.show()
```

```python
def perceptron(input_var,output_var):
    
    w1=w2=b=1
    lr=0.1
    
    for j in range(1000): #epochs
        
        for i in range(input_var.shape[0]): 
            
            z = w1*input_var[i][0] + w2*input_var[i][1] + b #f(x) = w1x1 + w2x2 + b
            
            if z*output_var[i] < 0 : # y.f(x) < 0
                w1 = w1 + lr*output_var[i]*input_var[i][0] #updating w1
                w2 = w2 + lr*output_var[i]*input_var[i][1] #updating w2
                b = b + lr*output_var[i] #updating bias
                
    return w1,w2,b
    
    
```

```python
w1,w2,b = perceptron(X,y)
```

```python
m = -(w1/w2)
c = -(b/w2)

print(m,c)
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + c

plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='viridis',s=100)
plt.ylim(-3,2)
plt.show()
```

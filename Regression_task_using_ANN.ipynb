```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Dropout,BatchNormalization
```

```python
import numpy as np
import pandas as pd
data = pd.read_csv('/content/boston.csv')
data.head()
```

```python
x = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x = sc.fit_transform(x)
pd.DataFrame(x).head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=101)
```

# DNN

```python
dnn = Sequential()

dnn.add(Dense(8,activation='relu',input_dim=13))
dnn.add(Dense(8,activation='relu'))
dnn.add(Dense(1,activation='linear'))

dnn.summary()
```

```python
# compile
dnn.compile(optimizer='sgd',loss='mean_squared_error')

#fit
history = dnn.fit(X_train,y_train,epochs=500)
```

```python
# Predict
y_pred = dnn.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
```
